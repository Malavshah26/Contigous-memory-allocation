```python
# Importing required modules in Python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
```

```python
# Loading the dataset
df = pd.read_csv("globalterrorismdb_0718dist.csv", low_memory=False, encoding='latin-1')
```

```python
# Checking for missing data
df.isnull().sum()
```

```python
# Checking the percentage of missing values
missing_percentage = df.isnull().sum()*100 /len(df)
print(missing_percentage)
```

```python
# Missing Values

null_cols = []
null_perc = []
size = len(df)
for col in df.columns:
    perc = df[col].isnull().sum()*100 / size
    if df[col].isnull().sum()>0:
        null_cols.append(col)
        null_perc.append(perc)
        
fig = plt.figure(figsize=(20, 30))
ax = fig.add_axes([0,0,1,1])
ax.barh(null_cols, null_perc)
for i, v in enumerate(null_perc):
    ax.text(v+1, i + .25, str(round(v, 2)) +'%', fontweight='bold')
ax.set_title('{} columns has missing value out of {}'.format(len(null_cols), len(df.columns)))
ax.set_ylabel("columns")
plt.show()
```

```python
print('{} columns has missing value out of {}'.format(len(null_cols), len(df.columns)))
```
